```python
import pandas as pd
from google.colab import files

# Upload the Excel file
uploaded = files.upload()

# Print the names of the uploaded files
print("Uploaded file(s):", uploaded.keys())

# Assuming you want to work with the first uploaded file
file_name = list(uploaded.keys())[0]

try:
    # Read the Excel file into a DataFrame
    data = pd.read_excel(file_name)

    # Display the DataFrame
    #print("DataFrame:")
    #print(data)
    # Print the first few rows of the DataFrame
    # print("First few rows of the DataFrame:")
    #print(data.head())

    # Select the entire DataFrame
    list_SLA = []

    # Create a file to store the results
    with open("formatted_results.txt", "w") as result_file:
        # Iterate over the rows
        for i in range(len(data)):
            # Selecting row i, column 3 (indexing starts from 0)
            value = data.iloc[i, 3]

            # Check if the value is not already in the list_SLA
            if value not in list_SLA:
                list_SLA.append(value)

        # Construct the desired result
        for k in list_SLA:
            list_booking = []  # Initialize list to store booking IDs
            for j in range(len(data)):
                if data.iloc[j, 3] == k:
                    booking = data.iloc[j, 4]  # Assuming booking column is at index 4
                    list_booking.append(booking)

            # Format the result for each SLA
            org_id = 43  # You can adjust the org_id accordingly
            booking_references = ','.join([f"'{ref}'" for ref in list_booking])
            formatted_result = f"SLA: {k} Corresponding Booking IDs: ORG_ID = {org_id} AND BOOKING_EXTERNAL_REFERENCE IN ({booking_references})"

            # Print the formatted result
            print(formatted_result)

            # Save the result to the file
            result_file.write(formatted_result + "\n")

    # Download the file
    files.download("formatted_results.txt")

except FileNotFoundError as e:
    print(f"Error: {e}")
```
